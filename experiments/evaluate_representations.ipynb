```python
import json

import ehrapy as ep
import pandas as pd
import numpy as np
import scanpy as sc
import patpy

import torch
import seaborn as sns
import matplotlib.pyplot as plt
from umap import UMAP
from scipy.spatial.distance import pdist, squareform

from sampleclr.contrastive_model import ContrastiveModel
from sampleclr.utils import get_sample_representations
```

```python
datasets = ["combat", "stephenson", "onek1k", "hlca"]
model_sizes = ["tiny", "middle", "large"]
model_types = ["SampleCLR_Unsupervised", "SampleCLR_Unsupervised_prior", "SampleCLR_supervised", "SampleCLR_supervised_prior"]
```

```python
def representation_to_distances(representation_df_path, model_name):
    repr_df = pd.read_csv(representation_df_path, index_col=0)
    # Remove duplicate sample indices (HLCA has a few)
    repr_df = repr_df[~repr_df.index.duplicated(keep="first")]

    print(f"Processing representation {model_name}")
    
    distances = pdist(repr_df, metric="cosine")
    
    distance_matrix = squareform(distances)
    distance_df = pd.DataFrame(
        distance_matrix,
        index=repr_df.index,
        columns=repr_df.index
    )

    return distance_df

def save_distances(distance_df, meta_adata, model_name):
    samples = meta_adata.obs_names
    samples = [sample for sample in samples if sample in distance_df.index]
    distance_df = distance_df.loc[samples][samples]
    
    print(f"Current number of samples: {len(samples)}")
    if len(samples) < meta_adata.n_obs:
        meta_adata = meta_adata[samples, :].copy()

    # It makes more sense to put distances to the obsp, but pp.neighbors expects it to be in obsm
    # So we put full distances matrix in obsm, and obsp will contain only the distances for nearest neighbors 
    meta_adata.obsm[f"{model_name}_distances"] = distance_df.loc[samples][samples].values
    print("Representation shape:", meta_adata.obsm[f"{model_name}_distances"].shape)
    meta_adata.uns["sample_representations"].append(model_name)
    return meta_adata


collapsed_representations = []
for dataset in datasets:
    representation_file_path = f"../data/{dataset}/{dataset}_representations.h5ad"
    meta_adata = ep.io.read_h5ad(representation_file_path)
    meta_adata.uns["sample_representations"] = list(meta_adata.uns["sample_representations"])
    with open(f"../configs/datasets/{dataset}.json", "r") as f:
        dataset_config = json.load(f)
    
    benchmark_schema = dataset_config["benchmark_schema"]

    model_size = "large" if dataset in ["hlca", "onek1k"] else "tiny"
    for model_type in model_types:
        # READ THE DATA
        representation_df_path = f"../results/{dataset}/{model_size}/{model_type}/{model_type}_representations.csv"
        model_name = f"{model_type}"
        
        try:
            distance_df = representation_to_distances(representation_df_path, model_name)
        except FileNotFoundError:
            continue

        low_distances_fraction = (distance_df < 0.01).mean().mean()
        print("Percentage of distances less than 0.01:", low_distances_fraction)
        if low_distances_fraction > 0.5:
            print("Representation is likely collapsed, skipping")
            collapsed_representations.append(f"{dataset}_{model_name}")
            continue

        # Remove samples that are not in the representation
        meta_adata = save_distances(distance_df, meta_adata, model_name)        

    # Add pascient
    representation_name = "pascient"
    representation_df_path = f"../results/{dataset}/pascient/patient_embeddings.csv"
    try:
        distance_df = representation_to_distances(representation_df_path, model_name=representation_name)
        meta_adata = save_distances(distance_df, meta_adata, model_name=representation_name)
    except FileNotFoundError as e:
        print("Error with pascient:", e)

    # Add mcbert
    representation_name = "mcbert"
    representation_df_path = f"../results/{dataset}/mcbert/sample_representations.csv"
    try:
        distance_df = pd.read_csv(representation_df_path, index_col=0)
        # Clean up sample names
        distance_df.columns = distance_df.index.str.replace("sample_adatas/", "")
        distance_df.index = distance_df.index.str.replace("sample_adatas/", "")
        meta_adata = save_distances(distance_df, meta_adata, model_name=representation_name)
    except FileNotFoundError as e:
        print("Error with mcbert:", e)
    
    # Fix variable type for Stephenson
    if dataset == "stephenson":
        meta_adata.obs[dataset_config["trajectory_variable"]].replace({
            "Healthy": 0,
            "Asymptomatic": 1,
            "Mild": 2,
            "Moderate": 3,
            "Severe": 4,
            "Critical": 5,
            "Critical ": 5,  # There is a typo in the data
            "Death": 6,
            "Non-covid": np.nan
        }, inplace=True)

    # Standartize the naming
    meta_adata.obsm["ehrapy_distances"] = meta_adata.obsp["ehrapy_neighbors_distances"].toarray()

    # EVALUATE THE REPRESENTATIONS
    model_names = sorted([x for x in meta_adata.uns["sample_representations"] if x.startswith("SampleCLR")])

    print("Putting neighbors to obsp")
    for representation_name in model_names + ["pascient", "mcbert"]:
        if f"{representation_name}_distances" not in meta_adata.obsm:
            print(f"No distances for {representation_name}, skipping")
            continue
        
        ep.pp.neighbors(
            meta_adata,
            use_rep=f"{representation_name}_distances",
            key_added=f"{representation_name}_neighbors",
            metric="precomputed"
        )

    print("Calculating leiden and UMAP")
    for representation_name in model_names + ["pascient", "mcbert"]:
        if f"{representation_name}_distances" not in meta_adata.obsm:
            print(f"No distances for {representation_name}, skipping")
            continue

        ep.tl.leiden(meta_adata, key_added=f"{representation_name}_leiden", neighbors_key=f"{representation_name}_neighbors")
        ep.tl.umap(meta_adata, neighbors_key=f"{representation_name}_neighbors")
        meta_adata.obsm[f"{representation_name}_umap"] = meta_adata.obsm["X_umap"]

    print("Resulting AnnData")
    print(meta_adata)

    np.random.seed(42)

    trajectory_metric_df = patpy.tl.evaluation.trajectory_correlation(meta_adata, dataset_config["root_sample"], dataset_config["trajectory_variable"], inverse_trajectory=dataset_config["inverse_trajectory"])
    trajectory_metric_df.sort_values("correlation", ascending=False, inplace=True)
    # trajectory_metric_df.to_csv(output_dir / "trajectory_metric.csv")

    plt.figure(figsize=(5, 10))
    sns.barplot(x="correlation", y=trajectory_metric_df.index, data=trajectory_metric_df, hue=trajectory_metric_df.index.str.contains("SampleCLR"), legend=False)
    plt.xlim(0, 1)
    plt.tight_layout()
    plt.title(dataset)
    plt.show()

    all_covariates = list(benchmark_schema["relevant"].keys()) + list(benchmark_schema["technical"].keys()) + list(benchmark_schema["contextual"].keys())
    for covariate in all_covariates:
        if covariate not in meta_adata.obs.columns:
            meta_adata.obs[covariate] = meta_adata[:, covariate].X.toarray().flatten()


    print("Benchmark schema:")
    print(benchmark_schema)

    knn_results = patpy.tl.evaluation.knn_prediction_score(meta_adata, benchmark_schema)

    avg_knn_score_df = pd.DataFrame(
        knn_results.groupby(["representation", "covariate_type"])["score"].aggregate(np.mean)
    ).reset_index()
    avg_knn_score_df = avg_knn_score_df.sort_values("score", ascending=False)
    # avg_knn_score_df.to_csv(output_dir / "avg_knn_score_df.csv")

    averaged_scores = avg_knn_score_df.pivot(index="representation", columns="covariate_type", values="score").sort_values("relevant", ascending=False)
    # averaged_scores.to_csv(output_dir / "averaged_scores.csv")

    plt.figure(figsize=(5, 5))
    sns.scatterplot(
        averaged_scores,
        x="relevant",
        y="technical",
        hue=averaged_scores.index.str.contains("SampleCLR"),
        legend=False
    )

    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.xlabel("Information retention")
    plt.ylabel("Batch mixing")
    plt.tight_layout()
    plt.title(dataset)
    plt.show()

    # Plot UMAP of representations
    # for model_name in model_names:    
    #     fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    #     umap_x = meta_adata.obsm[model_name + "_umap"][:, 0]
    #     umap_y = meta_adata.obsm[model_name + "_umap"][:, 1]
    #     trajectory_palette = "Reds_r" if dataset_config["inverse_trajectory"] else "Reds"

    #     sns.scatterplot(
    #         x=umap_x, y=umap_y,
    #         hue=meta_adata.obs[dataset_config["batch_variable"]], palette="tab10", s=100, ax=axes[0]
    #     )
    #     axes[0].legend(loc=(0, -.5), title=dataset_config["batch_variable"])

    #     sns.scatterplot(
    #         x=umap_x, y=umap_y,
    #         hue=meta_adata.obs[dataset_config["trajectory_variable"]], palette=trajectory_palette, s=100, ax=axes[1]
    #     )
    #     axes[1].legend(loc=(0, -.5), title=dataset_config["trajectory_variable"])

    #     sns.scatterplot(
    #         x=umap_x, y=umap_y,
    #         hue=meta_adata.obs[f"{model_name}_dpt_pseudotime"], palette="Reds", s=100, ax=axes[2]
    #     )
    #     axes[2].legend(loc=(0, -.5), title="DPT pseudotime")
    #     fig.suptitle(model_name)
    #     plt.show()

    averaged_scores = averaged_scores.join(trajectory_metric_df)
    averaged_scores["total"] = (averaged_scores["relevant"] + 0.5 * averaged_scores["technical"] + averaged_scores["correlation"]) / 2.5
    averaged_scores.sort_values("total", ascending=False, inplace=True)
    averaged_scores.to_csv(f"../results/{dataset}/spare_scores.csv")
```

```python
def parse_experiment_name(df):
    """Extract method, input space, distance metric, and aggregation method from the experiment name"""
    
    methods = ["pseudobulk", "grouped_pseudobulk", "random_vector", "cell_group_composition", "mrvi", "scpoli_sample_embeddings", "pilot", "mofa", "gloscope", "mrvi", "ehrapy", "random_vector",
    "SampleCLR_Unsupervised", "SampleCLR_Unsupervised_prior", "SampleCLR_supervised", "SampleCLR_supervised_prior",
    "multimil", "mixmil", "pascient", "mcbert"]
    for method in methods:
        # There is a collision between pseudobulk and grouped_pseudobulk,
        # but with the correct order this is fine
        rows_with_method = df[df["representation"].str.contains(method)]
        df.loc[rows_with_method.index, "method"] = method

    input_spaces = ["counts", "pca", "scpoli", "scvi_batch", "scvi_sample_id", "scanvi_batch", "scanvi_sample_id"]
    for input_space in input_spaces:
        rows_with_input_space = df[df["representation"].str.contains(input_space)]
        df.loc[rows_with_input_space.index, "input_space"] = input_space
    
    df.loc[df["representation"].str.contains("composition"), "input_space"] = "composition"
    df.loc[df["representation"] == "scpoli_sample_embeddings", "input_space"] = "counts"
    df.loc[df["representation"].str.contains("mrvi"), "input_space"] = "counts"
    df.loc[df["representation"].str.contains("ehrapy"), "input_space"] = "metadata"

    distance_metrics = ["euclidean", "cosine"]
    for distance_metric in distance_metrics:
        rows_with_distance_metric = df[df["representation"].str.contains(distance_metric)]
        df.loc[rows_with_distance_metric.index, "distance_metric"] = distance_metric
    df.loc[df["method"] == "SampleCLR", "distance_metric"] = "cosine"

    aggregation_methods = ["mean", "median", "sum"]
    for aggregation_method in aggregation_methods:
        rows_with_aggregation_method = df[df["representation"].str.contains(aggregation_method)]
        df.loc[rows_with_aggregation_method.index, "aggregation"] = aggregation_method

    df["supervised"] = False
    df.loc[df["representation"].str.contains("_supervised"), "supervised"] = True

    df["prior"] = False
    df.loc[df["representation"].str.contains("_prior"), "prior"] = True

    return df
```

```python
results = pd.DataFrame()
for dataset in datasets:
    dataset_scores = pd.read_csv(f"../results/{dataset}/spare_scores.csv")
    dataset_scores["dataset"] = dataset
    dataset_scores = parse_experiment_name(dataset_scores)
    results = pd.concat([results, dataset_scores])

results.reset_index(drop=True, inplace=True)
results = results.rename(columns={"correlation": "Trajectory preservation", "relevant": "Information retention", "technical": "Batch effect removal"})
```
