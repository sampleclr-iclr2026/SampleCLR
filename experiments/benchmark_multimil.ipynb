```python
import json
import gc

import anndata as ad
import multimil as mtm
import numpy as np
import scanpy as sc
import scvi
import warnings
import pandas as pd
from sklearn.metrics import classification_report
from pathlib import Path


warnings.filterwarnings("ignore")
```

```python
def get_supervised_task(dataset_config):
    """Extract the .obs column name for the supervised task from the dataset config"""
    task_dict = dataset_config["supervised_task"]
    if "classification" in task_dict:
        return task_dict["classification"][0]
    elif "regression" in task_dict:
        return task_dict["regression"][0]
    elif "ordinal_regression" in task_dict:
        return task_dict["ordinal_regression"][0]
    else:
        raise ValueError("No supervised task found in dataset config")
```

```python
datasets = ["combat", "stephenson", "onek1k", "hlca"]

for dataset in datasets:
    with open(f"../configs/datasets/{dataset}.json", "r") as f:
        dataset_config = json.load(f)

    task = get_supervised_task(dataset_config)

    scvi.settings.seed = 0
    print("Last run with scvi-tools version:", scvi.__version__)

    adata = sc.read_h5ad(dataset_config["data_path"])
    metadata = pd.read_csv(dataset_config["metadata_path"], index_col=0)
    print("Loaded adata:", adata.shape)
    print("Loaded metadata:", metadata.shape)

    layer = dataset_config["layer"]
    sample_key = dataset_config["sample_id_col"]

    print("Simplifying adata to only include the needed layer")
    adata = sc.AnnData(adata.obsm[layer], obs=adata.obs)
    print("adata:", adata)

    gc.collect()

    query_samples = metadata[metadata["split"] == "val"].index.tolist()

    query = adata[adata.obs[sample_key].isin(query_samples)].copy()
    adata = adata[~adata.obs[sample_key].isin(query_samples)].copy()

    query.obs["ref"] = "query"
    adata.obs["ref"] = "reference"

    print("Query samples (test set):", query.obs[sample_key].nunique())
    print("Reference samples (train set):", adata.obs[sample_key].nunique())

    classification_keys = dataset_config["supervised_task"].get("classification", [])

    regression_keys = dataset_config["supervised_task"].get("regression", [])
    print("Regression keys:", regression_keys)

    ordinal_regression_keys = dataset_config["supervised_task"].get("ordinal_regression", [])
    print("Ordinal regression keys:", ordinal_regression_keys)

    categorical_covariate_keys = classification_keys + ordinal_regression_keys + [sample_key]
    print("Categorical covariate keys:", categorical_covariate_keys)

    idx = adata.obs[sample_key].sort_values().index
    adata = adata[idx].copy()

    idx = query.obs[sample_key].sort_values().index
    query = query[idx].copy()

    print("Removing samples with no task label")
    adata = adata[adata.obs[task].notna()].copy()
    print("adata after removing samples with no task label:", adata.shape)

    try:
        mtm.model.MILClassifier.setup_anndata(
            adata,
            categorical_covariate_keys=categorical_covariate_keys,
            continuous_covariate_keys=regression_keys
        )

        mil = mtm.model.MILClassifier(
            adata,
            classification=classification_keys if len(classification_keys) else None,
            regression=regression_keys if len(regression_keys) else None,
            ordinal_regression=ordinal_regression_keys if len(ordinal_regression_keys) else None,
            sample_key=sample_key,
            class_loss_coef=0.1,
            regression_loss_coef=0.1
        )

        mil.train(lr=1e-3)
    except Exception as e:
        print(f"Error training model: {e}")
        continue

    mil.plot_losses()
    mil.get_model_output()

    query = query[query.obs[task].notna()].copy()
    print("Query after removing samples with no task label:", query.shape)
    
    new_model = mtm.model.MILClassifier.load_query_data(query, mil)
    new_model.get_model_output()

    print("Getting predictions for the test set")
    if "classification" in dataset_config["supervised_task"] or "ordinal_regression" in dataset_config["supervised_task"]:
        majority_voting_predictions = (query.obs
            .groupby(sample_key)[f"predicted_{task}"]
            .agg(lambda x: x.mode().iloc[0])
            )
    else:
        majority_voting_predictions = query.obs.groupby(sample_key)[task].mean()

    adata_both = ad.concat([adata, query])
    print("Saving sample representations")
    sample_reps = mtm.utils.get_sample_representations(adata_both, sample_key)
    sample_reps_df = pd.DataFrame(sample_reps.X, index=sample_reps.obs_names)

    print("Saving predictions")
    results_dir = Path(f"../results/{dataset_config['dataset_name']}/multimil/")
    results_dir.mkdir(parents=True, exist_ok=True)
    sample_reps_df.to_csv(results_dir / "sample_reps.csv")
    majority_voting_predictions.to_csv(results_dir / "val_set_predictions.csv") 
```
